```python

import pandas as pd
import random

def tallDF(categories, eventcount, values):
    eventlist = range(eventcount)
    categorycount = len(categories)
    eventIdList = [random.choice(eventlist) for _ in eventlist]
    result = pd.DataFrame (
        data = {
            "type": [ ":".join( [random.choice(categories[i]) for i in range(categorycount) ] ) for _ in eventlist ],
            "value" : [ random.choice(values) for _ in eventlist]
        }
    )
    return result
```

```python
taller = tallDF(categoryList, 10000,range(100))
```

```python
taller.head()
```

```python
def stragg(a):
    return ":".join([str(v) for v in a])
```

```python
pd.pivot_table(taller, index=taller.index, columns=taller["type"], aggfunc=stragg, dropna=False, margins=False)
```
